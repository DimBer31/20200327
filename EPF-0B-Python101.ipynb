## 4. functions, timing, plotting

```python
def volume(a,b,c):
    return a * b * c

print(volume(2,5,3))
```

```python
def factorial(n):
    if n<2:
        return 1
    else:
        return n*factorial(n-1)

print(factorial(100))
```
